```python
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
x = []
processy = []
lineary = []
thready = []
processz = []
linearz = []
threadz = []

for i,filename in enumerate(('logs/saida1P.txt', 'logs/saida2P.txt','logs/saidas.txt')):
    with open(filename, 'r') as f:
        txt = f.read()
    linear, thread, process = [], [], []
    for line in txt.split('\n'):
        if 'thread' in line:
            thread.append(float(re.findall(r'\d+[.]\d+', line)[0]))
        elif 'process' in line:
            process.append(float(re.findall(r'\d+[.]\d+', line)[0]))
        elif 'linear' in line:
            linear.append(float(re.findall(r'\d+[.]\d+', line)[0]))
    linear = np.asarray(linear)
    thread = np.asarray(thread)
    process = np.asarray(process)
    linear_avg,linear_std  = linear.mean(), linear.std()
    thread_avg,thread_std  = thread.mean(), thread.std()
    process_avg,process_std  = process.mean(), process.std()

    x.append(i)
    processy.append(process_avg)
    lineary.append(linear_avg)
    thready.append(thread_avg)
    
    processz.append(process_std)
    linearz.append(linear_std)
    threadz.append(thread_std)

processz = np.asarray(processz)
threadz = np.asarray(threadz)
linearz = np.asarray(linearz)

plt.errorbar(x, processy, processz**.5, label='process time')
plt.errorbar(x, lineary, linearz**.5, label='linear time',)
plt.errorbar(x,thready,threadz**.5, label='thread time')

plt.legend()
plt.show()
```
